```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pi = np.pi
# angles = [0, pi/4., pi/2., 3./4*pi, pi, 5./4*pi, 3./2*pi, 7./4*pi]
angles = np.linspace(0,2.*np.pi,20)
xs = np.sin(angles)
ys = np.cos(angles)
rs = [3,3.5,4]
```

```python
xs1 = xs*rs[0]
xs2 = xs*rs[1]
xs3 = xs*rs[2]

ys1 = ys*rs[0]
ys2 = ys*rs[1]
ys3 = ys*rs[2]
```

```python
plt.figure()
plt.plot(xs1, ys1, 'yo')
plt.plot(xs2, ys2, 'bo')
plt.plot(xs3, ys3, 'ro')
plt.axis('equal')
plt.show()
```

```python
rs1 = np.append(np.linspace(3,1,10), np.linspace(1,3,10))
rs2 = np.append(np.linspace(3.5,2,10), np.linspace(2,3.5,10))
rs3 = np.append(np.linspace(4,3,10), np.linspace(3,4,10))
```

```python
for t in range(20): 
    xs1 = xs*rs1[t]
    xs2 = xs*rs2[t]
    xs3 = xs*rs3[t]

    ys1 = ys*rs1[t]
    ys2 = ys*rs2[t]
    ys3 = ys*rs3[t]

    fig = plt.figure(figsize=(5,5))
    plt.plot(xs1, ys1, 'yo')
    plt.plot(xs2, ys2, 'bs')
    plt.plot(xs3, ys3, 'rv')
    plt.axis('off')
    plt.xlim(-5,5)
    plt.ylim(-5,5)
#     plt.axis('equal')
    plt.savefig(str(t)+'.jpg', bbox_inches='tight', pad_inches=0)
```
